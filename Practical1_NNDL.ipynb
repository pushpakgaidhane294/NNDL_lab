```python
import numpy as np
#Addition of two vector
vector1 = np.array([10,20,30])
vector2 = np.array([40,50,60])
print("vector 1 = ",vector1)
print("vector 2 = ",vector2)
result = vector1+vector2
print("Addition of two vector is ",result)
```

```python
#Matrix Multiplication
vector3 = np.array([[5,6],[2,3]])
vector4 = np.array([[4,7],[3,6]])
print("Vector 3 = ",vector3)
print("Vector 4 = ",vector4)
result1 = np.matmul(vector3,vector4)
print("Matrix Multiplication = ",result1)
```
